```python
import meshio
import numpy as np

med_path = "mesh/cantilever_beam.med"
xdmf_path = "mesh/mesh.xdmf"

mesh = meshio.read(med_path)
points, cell, cell_data, field_data = mesh.points, mesh.cells, mesh.cell_data, mesh.field_data
a = cell['tetra']
meshio.write(xdmf_path, meshio.Mesh(
    points=points[:,:],
    cells={"tetra": a},
    ))
print("Converted .med to .xdmf location:" + xdmf_path)
```
